```python
def get_pyramid_dataset(grid_size = 4, size = 3):
    if type(grid_size) == tuple:
        dims = grid_size
    elif type(grid_size) == int:
        dims = (grid_size, grid_size)
        
    mats = []
    
    for x_start, y_start, in np.ndindex(grid_size - size + 1, grid_size - size + 2):
        mat = np.ones(dims) * w
        
        level_size = size
        x = x_start
        y = y_start

        while level_size > 0:
            mat[grid_size - y - 1][x: (x + level_size)] = [b] * (level_size)

            level_size -= 2
            x += 1
            y += 1

        mats.append(mat)
    
    return np.array(mats)


            

def get_ship_dataset(grid_size = 4, size = 3):
    if type(grid_size) == tuple:
        dims = grid_size
    elif type(grid_size) == int:
        dims = (grid_size, grid_size)
    
    mats = []
    
    for x_start, y_start, in np.ndindex(grid_size - size + 1, grid_size - size + 1):

        mat = np.ones(dims) * w
        x = x_start
        y = y_start

        for i in range(size):
            if i % 2 == 0:
                mat[y][x:(x+size-1)] = [b] * (size - 1)
            else:
                mat[y][(x + size - 2):(x + size)] = [b] * 2

            y += 1

        mats.append(mat)
    
    return np.array(mats)

def get_box_dataset(grid_size=4, size=3):
    if type(grid_size) == tuple:
        dims = grid_size
    elif type(grid_size) == int:
        dims = (grid_size, grid_size)
    
    mats = []
    
    for x_start, y_start, in np.ndindex(grid_size - size + 1, grid_size - size + 1):

        mat = np.ones(dims) * w
        x = x_start
        y = y_start

        for i in range(size):
            if i == 0 or i == size - 1:
                mat[y][x:(x+size)] = [b] * size
            else:
                mat[y][x] = b
                mat[y][x + size -  1] = b

            y += 1

        mats.append(mat)
    
    return np.array(mats)
```
